```python
import random
import pandas as pd
from datetime import datetime, timedelta
import time
```

```python
def generate_pengiriman(n):
    data = []
    base_time = datetime.now()

    for i in range(n):
        data.append({
            "id_paket": f"PKT{i+1:05d}",
            "kurir": random.choice(["JNE", "J&T", "SiCepat", "AnterAja"]),
            "asal": random.choice(["Jakarta", "Bandung", "Surabaya"]),
            "tujuan": random.choice(["Purwokerto", "Semarang", "Yogyakarta"]),
            "jenis_layanan": random.choice(["Reguler", "Ekspres"]),
            "status_pengiriman": random.choice(["Diproses", "Dikirim", "Sampai"]),
            "waktu_pengiriman": base_time + timedelta(minutes=random.randint(0, 100000))
        })
    return data
```

```python
data_awal = generate_pengiriman(1000)
df_awal = pd.DataFrame(data_awal)
df_awal
```

```python
def bubble_sort_iterative(data):
    arr = data.copy()
    n = len(arr)

    for i in range(n):
        for j in range(0, n - i - 1):
            if arr[j]["waktu_pengiriman"] > arr[j + 1]["waktu_pengiriman"]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    return arr
```

```python
def bubble_sort_recursive(arr, n=None):
    if n is None:
        arr = arr.copy()
        n = len(arr)

    if n == 1:
        return arr

    for i in range(n - 1):
        if arr[i]["waktu_pengiriman"] > arr[i + 1]["waktu_pengiriman"]:
            arr[i], arr[i + 1] = arr[i + 1], arr[i]

    return bubble_sort_recursive(arr, n - 1)
```

```python
def merge_sort_recursive(arr):
    if len(arr) <= 1:
        return arr

    mid = len(arr) // 2
    left = merge_sort_recursive(arr[:mid])
    right = merge_sort_recursive(arr[mid:])

    return merge(left, right)

def merge(left, right):
    result = []
    i = j = 0

    while i < len(left) and j < len(right):
        if left[i]["waktu_pengiriman"] <= right[j]["waktu_pengiriman"]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1

    result.extend(left[i:])
    result.extend(right[j:])
    return result
```

```python
def merge_sort_iterative(arr):
    width = 1
    n = len(arr)
    result = arr.copy()

    while width < n:
        for i in range(0, n, 2 * width):
            left = result[i:i + width]
            right = result[i + width:i + 2 * width]
            result[i:i + 2 * width] = merge(left, right)
        width *= 2

    return result
```

```python
sorted_example = merge_sort_iterative(data_awal)
pd.DataFrame(sorted_example)
```

```python
input_sizes = [10, 50, 100, 500, 1000]
runs = 3

results = {
    "Bubble Iteratif": [],
    "Bubble Rekursif": [],
    "Merge Iteratif": [],
    "Merge Rekursif": []
}

for n in input_sizes:
    for algo in results:
        total_time = 0
        for _ in range(runs):
            data = generate_pengiriman(n)

            start = time.time()
            if algo == "Bubble Iteratif":
                bubble_sort_iterative(data)
            elif algo == "Bubble Rekursif":
                bubble_sort_recursive(data)
            elif algo == "Merge Iteratif":
                merge_sort_iterative(data)
            else:
                merge_sort_recursive(data)

            total_time += time.time() - start

        results[algo].append(total_time / runs)
```

```python
pd.options.display.float_format = '{:.6e}'.format

df_hasil = pd.DataFrame({
    "Jumlah Paket": input_sizes,
    "Bubble Sort Iteratif": results["Bubble Iteratif"],
    "Bubble Sort Rekursif": results["Bubble Rekursif"],
    "Merge Sort Iteratif": results["Merge Iteratif"],
    "Merge Sort Rekursif": results["Merge Rekursif"]
})

df_hasil
```

```python
import time

def ukur_waktu(sort_func, data, run=3):
    total = 0
    for _ in range(run):
        copy_data = data.copy()
        start = time.time()
        sort_func(copy_data)
        end = time.time()
        total += (end - start)
    return total / run


df_waktu = pd.DataFrame({
    "Algoritma": [
        "Merge Sort (Iteratif)",
        "Merge Sort (Rekursif)",
        "Bubble Sort (Iteratif)",
        "Bubble Sort (Rekursif)"
    ],
    "Rata-rata Waktu (detik)": [
        ukur_waktu(merge_sort_iterative, data),
        ukur_waktu(merge_sort_recursive, data),
        ukur_waktu(bubble_sort_iterative, data),
        ukur_waktu(bubble_sort_recursive, data)
    ]
})

df_waktu
```
